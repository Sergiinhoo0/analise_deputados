```python
import pandas as pd
```

```python
arquivo = "dados_brutos_23-05-22.xlsx"
```

```python
df_raw = pd.read_excel(arquivo)
```

```python
df_raw.head()
```

```python
df1 = df_raw.drop("Votação", axis = 1)
```

```python
df1.head()
```

```python
pautas = df1.drop("Parlamentar", axis = 1).columns
```

```python
pautas_simp = []
for p in range(1, len(pautas)+1):
    pautas_simp.append("V%i" %p)
```

```python
pautas = pd.DataFrame(pautas, index = pautas_simp, columns = ["Votação"])
```

```python
pautas.head()
```

```python
colunas_simp = pautas_simp.copy()
colunas_simp.insert(0, "Parlamentar")
```

```python
df1.columns = colunas_simp
```

```python
df1.head()
```

```python
df2 = df1.copy()

df2 = df2.replace("-", False)
df2 = df2.replace("Não votou", False)
```

```python
df2.head()
```

```python
contador = 1
weight = 0
parlamentares = df1["Parlamentar"]
df_rede_vot = None
df_rede_vot = pd.DataFrame(columns = ["source", "target", "weight"])


for i in parlamentares:
    for j in parlamentares:
        if (df2[df2["Parlamentar"] == i].index[0] < df2[df2["Parlamentar"] == j].index[0]):
            
            df2_1 = (df2[df2["Parlamentar"] == i].transpose())
            df2_1 = df2_1.rename(columns=df2_1.iloc[0]).iloc[1:]
            df2_2 = (df2[df2["Parlamentar"] == j].transpose())
            df2_2 = df2_2.rename(columns=df2_2.iloc[0]).iloc[1:]
            
            df2_3 = df2_1.join(df2_2)
            df2_3 = df2_3.replace(["Sim", "Não", "Abstenção", "Obstrução", "Eleição" ], True)
            
            div = (df2_3.iloc[ : , 0] == df2_3.iloc[ : , 1]).sum()
            
            weight = ( (df2[df2["Parlamentar"] == i].squeeze() ) == ( df2[df2["Parlamentar"] == j].squeeze()) ).sum()
 
            if (weight > 0):
                df_rede_vot = df_rede_vot.append({"source":i, "target":j, "weight":weight, "total_vot":div}, ignore_index = True)
            
    print(contador)
    contador += 1
```

```python
df_rede_vot.head()
```

```python
df_rede_vot.to_csv("df_rede_vot.csv", encoding = "utf-8", index = False, sep = ";")
```

```python
edge_list_vot = df_rede_vot.copy()
edge_list_vot["weight"] = edge_list_vot["weight"] / edge_list_vot["total_vot"]
edge_list_vot.drop("total_vot", axis = 1, inplace = True)
```

```python
edge_list_vot.head()
```

```python
edge_list_vot.to_csv("edge_list_vot.csv", encoding = "utf-8", index = False, sep = ";")
```
